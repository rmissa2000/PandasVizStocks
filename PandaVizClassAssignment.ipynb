```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

Above are the nessecary packages for the following financial analytics. We will be using Amazon (AMZN). 

```python
start = datetime.datetime(2022, 1, 1)
end = datetime.datetime(2022, 2, 15)
```

```python
AMZN = data.DataReader("AMZN", 'yahoo', start, end)
```

```python
AMZN
```

We need to reset the uploaded data on AMZN to adjust indexes for analysis. 

```python
myAMZN = AMZN.reset_index()
```

```python
myAMZN
```

Below is a time series analysis of AMZN's volume for 2022 YTD.

```python
plt.style.use('ggplot')
```

```python
plt.style.use('ggplot')
myAMZN.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

Barchart for opening price for Dec 2021

```python
start1 = datetime.datetime(2021, 12, 1)
end1 = datetime.datetime(2021, 12, 31)
newAMZN = data.DataReader("AMZN", 'yahoo', start1, end1)
newAMZN1 = newAMZN.reset_index()
newAMZN["Open"].plot.bar()
```

Now we will 'analyze' Microsoft (MSFT)

```python
MSFT = data.DataReader("MSFT", 'yahoo', start, end)
myMSFT = MSFT.reset_index()
myMSFT
```

Time series analysis of volume for YTD.

```python
plt.style.use('bmh')
myMSFT.plot.line(x='Date',y='Volume',figsize=(12,3),lw=1)
```

To do the density plot, we need to subset the MSFT set to only contain the open and close values. We will then display the density plot.

```python
OpenCloseMSFT = myMSFT[["Open", "Close" ]]
OpenCloseMSFT
```

```python
OpenCloseMSFT.plot.density()
```
